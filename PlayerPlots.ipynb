# Plotting the Shift Ban Effect By Player

## 1. Importing and setting up the data frames ##

```python
import numpy as np
import matplotlib as mlp
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv(r'C:\Users\theod\OneDrive\Desktop\MLB Baseball Analytics 2023\1 - Shift Effect\Cleaned Data\luck_increase.csv')

#setting up X and Y axes
X = df['gb_luck_increase']
Y = df['ld_luck_increase']

#setting up points to be annotated
luckiest_df = df.nlargest(5,'total_luck_increase')
lucky_ids = luckiest_df.index

unluckiest_df = df.nsmallest(5,'total_luck_increase')
unlucky_ids = unluckiest_df.index
```

## 2. Creating the scatterplot to show the luck increase spread
* X-axis is ground ball luck
* Y-axis is line drive luck
* Color scale shows weighted overall luck increase
* Top and bottom 5 players are labelled

```python
### USING PYPLOT FUNCTIONS TO MAKE A SCATTERPLOT

plt.axvline(0,ls='--',lw = 1)
plt.axhline(0,ls='--',lw = 1) #creating lines for axes
plt.style.use('seaborn') #styling
plt.scatter(X,Y, c=df.total_luck_increase, cmap = 'coolwarm', edgecolors='k') #creating scatterplot
plt.colorbar(label = 'Overall Luck Increase') #creating key for colors
plt.ylim(-0.35,0.35)
plt.xlim(-0.35,0.35) #setting size of view
plt.title('Luck Increase Resulting from the Shift Ban') 
plt.xlabel('Ground Ball Luck Increase')
plt.ylabel('Line Drive Luck Increase') #title and axes labels

#labelling top and bottom 5 hitters
for index in lucky_ids:
    plt.annotate(luckiest_df.player_name[index],(luckiest_df.gb_luck_increase[index],luckiest_df.ld_luck_increase[index]))

for index in unlucky_ids:
    plt.annotate(unluckiest_df.player_name[index],(unluckiest_df.gb_luck_increase[index],unluckiest_df.ld_luck_increase[index]))
```
